# Preprocessing
We preprocess the data for training word2vec model.

```python
import pandas as pd
from tqdm import tqdm
import re
import string
```

```python
# Load the dataset.

PATH = '/content/drive/MyDrive/Datasets/kaggle_rock.csv'
df = pd.read_csv(PATH)
df = df['lyrics']
df.head()
```

```python
# Since we want to train embeddings, we can flatten 
# the dataset. One line would therefore be one context.

ifile = open(PATH, 'r')
ifile.readline()  # csv header
# We'll save flattened lyrics for later use.
ofile = open('/content/drive/MyDrive/Datasets/flat.txt', 'w')
legal = string.ascii_lowercase + '\n' + "' "
legal  # Allowed characters.
```

```python
# Example of flattened lyrics.

f = filter(lambda c: c in legal, df[0])
s = ''.join(f)
print(s)
```

```python
# Go through the dataset, filtering the lyrics, and save the result.
for line in tqdm(ifile.readlines()):
    f = filter(lambda c: c in legal, line)
    ofile.write(''.join(f))
```

```python
ifile.close()
ofile.close()
```
